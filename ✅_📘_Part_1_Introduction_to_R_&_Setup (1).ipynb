# **üè∑Ô∏è 5Ô∏è‚É£ Variables and Data Types (Quick Intro)**

# üìå Variables store information.

- In R, most numbers are stored as "numeric".
- This is a general term that usually means "double precision" (numbers with decimals).
- R doesn‚Äôt separate whole numbers (integers) and decimals (floats) by default like some other languages do.
- So even if you write age <- 25, R still stores it as a "numeric" (a decimal number internally).

```r
# You can create an actual integer by adding an L, like this:
age_integer <- 25L
print(class(age_integer))  # This will show "integer"
```

```r
# Numeric: used for numbers (whole or with decimals)
age <- 25
height <- 5.8

# Character (text): used for words or text values
name <- "Bob"

# Logical: used for TRUE or FALSE values
is_student <- TRUE

# Print the values
print(name)
print(class(name))       # Shows the "type" of data stored in 'name' ‚Äî here it's "character"

print(age)
print(class(age))        # "numeric" means it's a number you can do math with

print(height)
print(class(height))     # Also "numeric", because it‚Äôs a number with a decimal point

print(is_student)
print(class(is_student)) # "logical" means TRUE/FALSE data (yes/no type)
```
